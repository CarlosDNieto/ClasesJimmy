# NoSQL

* **Stands for Not Only SQL**


* **Non relational database (No Tables)**


* **Flexible with bigData and real time apps**


* **Multiple types of NoSQL Databases**

---

## What is BigData?

* Un término para los conjuntos de datos que son tan grandes que los métodos tradicionales de almacenamiento y procesamiento son inadecuados


* Se debe al incremento masivo de datos en los últimos años


* Gracias a las redes sociales, motores de búsqueda, etc


* Representa retos en almacenaje, procesamiento, captura, análisis, transferencias, etc

---

## Ventajas de NoSQL vs RDBMS

* Se encarga de manejar BidData


* No tiene una estructura definida predeterminadamente, esto da flexibilidad


* Se puede encargar de datos no estructurados


* Más baratas de manejar


* Escalamiento horizontal y vertical

## Ventajas de RDBMS sobre NoSQL

* Mejor en datos relacionados


* Normalización, organiza datos de manera que elemina redundancia de datos


* Lenguaje conocido SQL


* Integridad de datos 


* ACID Compliance - atomicity, consistency, isolation, durability


---

## Tipos de bases de datos NoSQL

- Document Databases [MongoDB, CouchDB]


- Column Databases [Apache Cassandra]


- Key-value stores [Redis, Couchbase Server]


- Cache Systems [Redis, Memcache]


- Graph Databases [Neo4J]
